```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('C:/Users/Dell/testset_edited.csv', parse_dates=['datetime_utc']) #Master is the cleaned dataset that Gautam gave
df.head()
```

```python
df.columns = df.columns.to_series().apply(lambda x: x.strip()) # Stripping any extra spaces in the column names
```

```python
df_actual1 =  df[['datetime_utc','_tempm']]
df_actual1.head(2)
```

```python
df_actual1.columns = ['Date','Temperature'] 
df_actual1.head()
```

```python
df1=df_actual1.groupby("Date")["Temperature"].mean()
```

```python
df1.to_csv(r'C:/Users/Dell/Actual.csv',header= False)
```

```python
df_actual= pd.read_csv('C:/Users/Dell/Actual.csv')
df_actual.head(2)
```

```python
df_actual.columns = ['Date','Temperature'] 
df_actual.head()
```

```python
df_predicted=pd.read_csv('Predicted.csv')#Data that sachin and meghana gave
df_predicted.head()
```

```python
df_predicted.columns = ['Date','Temperature'] 
df_predicted.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_predicted["Date"], y=df_predicted['Temperature'], name="Actual",
                         line_color='deepskyblue'))

fig.add_trace(go.Scatter(x=df_actual["Date"], y=df_actual['Temperature'], name="Predicted",
                         line_color='dimgray'))

fig.update_layout(title_text='Interactive Weather',
                  xaxis_rangeslider_visible=True)
fig.show()
```
